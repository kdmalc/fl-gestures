Running KNN (treating each time step as its own sample)

```python
k_range=range(1, 21)

knn_accuraciesHOT = []
knn_accuraciesHOG = []
knn_accuraciesHOU = []

for k in k_range:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_2Dfeatures_train_embeddings, train_2D_labels)
    
    y_pred = knn.predict(X_2Dfeatures_test_HOTembeddings)
    accuracy = accuracy_score(testHOT_2D_labels, y_pred)
    knn_accuraciesHOT.append(accuracy)
    print(f'k={k}, HOT Accuracy: {accuracy*100:.2f}%')

    y_pred = knn.predict(X_2Dfeatures_test_HOGembeddings)
    accuracy = accuracy_score(testHOG_2D_labels, y_pred)
    knn_accuraciesHOG.append(accuracy)
    print(f'k={k}, HOG Accuracy: {accuracy*100:.2f}%')

    y_pred = knn.predict(X_2Dfeatures_test_HOUembeddings)
    accuracy = accuracy_score(testHOU_2D_labels, y_pred)
    knn_accuraciesHOU.append(accuracy)
    print(f'k={k}, HOU Accuracy: {accuracy*100:.2f}%\n')

# Plot the accuracies for each k
plt.plot(k_range, knn_accuraciesHOT, marker='o', label='Held Out Trial')
plt.plot(k_range, knn_accuraciesHOG, marker='o', label='Held Out Gestures')
plt.plot(k_range, knn_accuraciesHOU, marker='o', label='Held Out Users')
plt.xlabel('Number of Neighbors (k)')
plt.ylabel('Accuracy')
plt.title('KNN Accuracy vs Number of Neighbors')
plt.legend()
plt.grid(True)
plt.show()
```

## SVC

```python
from sklearn.model_selection import GridSearchCV
```

```python
#def svm_hyperparameter_tuning(X_train, y_train, X_test, y_test, kernels=['linear', 'rbf', 'poly', 'sigmoid'], use_condensed_grid=True):
kernels=['linear', 'rbf', 'poly', 'sigmoid']

print("Starting\n")

# FULL
C_lst = [0.1, 1, 10, 100]
gamma_lst = [1, 0.1, 0.01, 0.001]
param_grid = {
    'linear': {'C': C_lst},
    'rbf': {'C': C_lst, 'gamma': gamma_lst},
    'poly': {'C': C_lst, 'degree': [2, 3, 4], 'gamma': gamma_lst, 'coef0': [0, 1, 10]},
    'sigmoid': {'C': C_lst, 'gamma': gamma_lst, 'coef0': [0, 1, 10]}
}
    
best_params = {}
best_accuracies = {}

svc_accuraciesHOT = []
svc_accuraciesHOG = []
svc_accuraciesHOU = []

for kernel in kernels:
    grid = GridSearchCV(SVC(kernel=kernel, random_state=42), param_grid[kernel], refit=True, verbose=0)
    grid.fit(X_2Dgestures_train_embeddings, train_labels)
    
    y_pred = grid.predict(X_2Dgestures_test_HOTembeddings)
    accuracy = accuracy_score(testHOT_labels, y_pred)
    svc_accuraciesHOT.append(accuracy)
    print(f'HOT Accuracy: {accuracy*100:.2f}%')

    y_pred = grid.predict(X_2Dgestures_test_HOGembeddings)
    accuracy = accuracy_score(testHOG_labels, y_pred)
    svc_accuraciesHOG.append(accuracy)
    print(f'HOG Accuracy: {accuracy*100:.2f}%')

    y_pred = grid.predict(X_2Dgestures_test_HOUembeddings)
    accuracy = accuracy_score(testHOU_labels, y_pred)
    svc_accuraciesHOU.append(accuracy)
    print(f'HOU Accuracy: {accuracy*100:.2f}%\n')

    avg_accuracy = (svc_accuraciesHOT[-1] + svc_accuraciesHOG[-1] + svc_accuraciesHOU[-1])/3

    best_params[kernel] = grid.best_params_
    best_accuracies[kernel] = avg_accuracy
    print(f'Kernel={kernel}, Best Params: {grid.best_params_}, Accuracy: {avg_accuracy*100:.2f}%')
    #print(f"Classification Report for {kernel} kernel:\n", classification_report(y_test, y_pred))

plt.bar(best_accuracies.keys(), best_accuracies.values())
plt.xlabel('Kernel Type')
plt.ylabel('Accuracy')
plt.title('SVM Accuracy vs Kernel Type')
#plt.grid(True)
plt.show()
```
