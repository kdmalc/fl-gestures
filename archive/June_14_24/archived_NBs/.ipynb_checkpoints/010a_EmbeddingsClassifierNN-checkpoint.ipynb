- One Hot Encoding: This representation is often used in multi-label classification where an instance can belong to multiple classes simultaneously.
- Integer Encoding: This is the preferred representation for multi-class classification where each instance belongs to exactly one class.
    - In PyTorch, when using nn.CrossEntropyLoss, the labels should be integer-encoded and not one-hot encoded. nn.CrossEntropyLoss internally applies a softmax to the model's raw output logits and then computes the loss, expecting the target labels as integers.
 
> The output layer of your network should have a size equal to the number of classes.
For example, if you have 28 unique gesture classes, the output layer should have 28 units.
