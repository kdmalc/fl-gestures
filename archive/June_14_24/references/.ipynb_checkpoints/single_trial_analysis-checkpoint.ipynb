# filtered EMG signal 1

40 Hz HP filter, rectify, 40 Hz LP filter

```python
def rectify_EMG(data):
    # 0. demean EMG data
    data = data - np.mean(data)
    
    # 1. high-pass filter @ 40 hz, 4th order butterworth
    sos_high = signal.butter(N=4,Wn=40,btype='high',fs=2000,output='sos')
    hp_filtered = signal.sosfilt(sos_high,data)
    
    # 2. rectify
    rectified = abs(hp_filtered)
    
    # 3. low-pass filter @ 40 Hz, 4th order butterworth
    sos_low = signal.butter(N=4,Wn=40,btype='low',fs=2000,output='sos')
    lp_filtered = signal.sosfilt(sos_low,rectified)

    # sos_high = signal.butter(N=2,Wn=40,btype='high',fs=2000,output='sos')
    # hp_filtered = signal.sosfiltfilt(sos_high,data)
    
    # rectified = abs(hp_filtered)
    
    # sos_low = signal.butter(N=2,Wn=40,btype='low',fs=2000,output='sos')
    # lp_filtered = signal.sosfiltfilt(sos_low,rectified)
    return lp_filtered

df_EMG_filt = df_EMG.apply(rectify_EMG,axis=0)

fig,axs = plt.subplots(4,4,figsize=(20,10),sharex=True, sharey=False)
ch=0
for ix in range(4):
    for iy in range(4):
        axs[ix,iy].plot(time_EMG,df_EMG[ch]-np.mean(df_EMG[ch]),color='#9CDDF9')
        axs[ix,iy].plot(time_EMG,df_EMG_filt[ch])
        axs[ix,iy].set_title("electrode: "+muscles[ch])# +str(ch+1))
        ch+=1
```

# filtered EMG signal 2 

30-450 Hz bandpass, and then rectify

```python
def rectify_EMG(data):
    # 0. demean EMG data
    data = data - np.mean(data)
    
    # 1. bp filter @ 30-450 hz, 4th order butterworth
    sos_bp = signal.butter(N=4,Wn=[50,150],btype='bandpass',fs=2000,output='sos')
    hp_bp = signal.sosfilt(sos_bp,data)
    
    # 2. rectify
    rectified = abs(hp_bp)
    return rectified

df_EMG_filt = df_EMG.apply(rectify_EMG,axis=0)

fig,axs = plt.subplots(4,4,figsize=(20,10),sharex=True, sharey=False)
ch=0
for ix in range(4):
    for iy in range(4):
        axs[ix,iy].plot(time_EMG,df_EMG_filt[ch])
        axs[ix,iy].set_title("electrode: "+muscles[ch]+str(ch+1))
        ch+=1
```

```python
def movavg_EMG(data,window=100,overlap=0.5,fs=2000):
    # 200 ms window, overlap = 0 is no overlap, overlap=1 is complete overlap 
    N = int(1*window/1000*fs) # number of datapoints in one window
    N_overlap = int(N*overlap)
    
    movavg_data = []
    
    ix = N
    while ix < len(data):
        movavg_data.append(np.mean(data[ix-N:ix]))
        ix = ix + (N - N_overlap)
    
    return np.asarray(movavg_data)
```

```python
df_EMG_movavg = df_EMG_filt.apply(movavg_EMG,axis=0)
time_movavg = np.linspace(0,time_EMG[-1],len(df_EMG_movavg))
fig,axs = plt.subplots(4,4,figsize=(10,5),sharex=True, sharey=False)
ch=0
for ix in range(4):
    for iy in range(4):
        axs[ix,iy].plot(time_movavg,df_EMG_movavg[ch])
        axs[ix,iy].set_title(muscles[ch])#+str(ch+1))
        ch+=1
        if iy == 0:
            axs[ix,iy].set_ylabel('voltage (V)')
        if ix==3:
            axs[ix,iy].set_xlabel('time (sec)')
        axs[ix,iy].spines['top'].set_color('black')
        axs[ix,iy].spines['right'].set_color('black')
plt.tight_layout()
```

```python
# dead reckoning 
start = 10
omegas = r_wrist_IMU.iloc[:,3:].to_numpy() 
accels = r_wrist_IMU.iloc[:,:3].to_numpy()
omegas = omegas - np.mean(omegas[:start,:],axis=0)
```

https://towardsdatascience.com/dead-reckoning-is-still-alive-8d8264f7bdee#:~:text=One%20of%20the%20common%20algorithms,the%20GPS%20receivers%20or%20cameras.

```python
def R(angles):
    theta, gamma, psi = angles
    r11 = np.cos(theta)*np.cos(psi)
    r12 = np.sin(gamma)*np.sin(theta)*np.cos(psi) - np.cos(gamma)*np.sin(psi)
    r13 = np.cos(gamma)*np.sin(theta)*np.cos(psi) + np.sin(gamma)*np.sin(psi)
    r21 = np.cos(theta)*np.sin(psi)
    r22 = np.sin(gamma)*np.sin(theta)*np.sin(psi) + np.cos(gamma)*np.cos(psi)
    r23 = np.cos(gamma)*np.sin(theta)*np.sin(psi) - np.sin(gamma)*np.cos(psi)
    r31 = -np.sin(theta)
    r32 = np.sin(gamma)*np.cos(theta)
    r33 = np.cos(gamma)*np.cos(theta)
    return np.array([[r11,r12,r13],[r21,r22,r23],[r31,r32,r33]])

def cost(angles,a):
    g_ = np.reshape(np.array([0,0,-g]),[3,1]) # gravity when person is at rest
    a = np.reshape(a,[3,1]) # reshape a to a column vector
    return np.linalg.norm(-g_+R(angles)@a)

def get_R_init(a): 
    # a is acceleration
    out = minimize(lambda angles: cost(angles,a),.1*np.ones((3,)))
    return R(out.x)
```

```python
# take avg of 1st 20 datapoints
accels_avg = np.mean(r_wrist_IMU.iloc[:10,:3].to_numpy(),axis=0)
print(accels_avg)
R_0 = get_R_init(accels_avg)
R_0
```

```python
R_0@accels_avg - np.array([0,0,-g])# this is correct
```

```python
def get_R_next(R,omega):
    omega = omega / 180 * np.pi # convert to radians
    x,y,z = omega
    Omega = np.array([[0, -z, y],[z,0,-x],[-y,x,0]])
    return R@Omega
def get_v_next(R,a):
    g_ = np.reshape(np.array([0,0,-g]),[3,1])
    a = np.reshape(a,[3,1]) # reshape a to a column vector
    return R@a-g_

# take avg of 1st 10 datapoints
accels_avg = np.mean(r_wrist_IMU.iloc[:start,:3].to_numpy(),axis=0)
Rprev = get_R_init(accels_avg)
ps = []
vs = []
ps.append(np.zeros((3,)))
vs.append(np.zeros((3,)))
for a in accels[start:]:
    ix = start
    Rnext = get_R_next(Rprev,omegas[ix])# get next R
    vnext = get_v_next(Rnext,a)# get next v
#     print(vnext)
    vs.append(vs[-1]+np.squeeze(vnext)*1/f_IMU)
#     print(vs)
    ps.append(ps[-1]+vs[-1]*1/f_IMU)
#     print(ps)
    Rprev = Rnext
    ix += 1
#     break;
ps=np.asarray(ps)
vs = np.asarray(vs)
```
